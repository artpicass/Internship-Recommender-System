**SVD (Singular Value Decomposition )**

```
final_df = ratings.pivot(index='UserID',columns='MovieID',values='Rating')
```

```
final_df.fillna(0,inplace=True)
final_df
```

```
rel = final_df.values
meann = np.mean(rel, axis = 1)
final_mean = rel - meann.reshape(-1, 1)
```

```
from scipy.sparse.linalg import svds
U, sig, Vt = svds(final_mean, k = 50)
```

```
sig = np.diag(sig)
```

```
all_user = np.dot(np.dot(U, sig), Vt) + meann.reshape(-1, 1)
```

```
pred_df = pd.DataFrame(all_user, columns = final_df.columns)
pred_df.head()
```

```
pred_df.shape
```

```
def recommend_movies(userID, n=10):
    
    # Get and sort the user's predictions
    user_idx = userID - 1 # UserID starts at 1, not 0
    user_pred = pred_df.iloc[user_idx].sort_values(ascending=False) # UserID starts at 1
    
    # Get the user's data and merge in the movie information.
    user_data = ratings[ratings.UserID == (userID)]
    user_full = (user_data.merge(movies, how = 'left', left_on = 'MovieID', right_on = 'MovieID').sort_values(['Rating'], ascending=False))
    # Recommend the highest predicted rating movies that the user hasn't seen yet.
    recommendations = (movies[~movies['MovieID'].isin(user_full['MovieID'])].merge(pd.DataFrame(user_pred).reset_index(), how = 'left', left_on = 'MovieID', right_on = 'MovieID').rename(columns = {user_idx: 'Predictions'}).sort_values('Predictions', ascending = False).iloc[:n, :-1])
    
    print ("User {0} has already rated {1} movies.".format(userID, user_full.shape[0]))
    print ("Recommending highest {0} predicted rating movies not already rated.".format(n))
    pd.set_option('display.max_columns', None)
    return(recommendations)
```

```
from collections import defaultdict

from surprise import SVD
from surprise import Dataset
```

```
def get_top_n(predictions, n=10):
# First map the predictions to each user.
    top_n = defaultdict(list)
    for UserID, MovieID, true_r, est, _ in predictions:
        top_n[UserID].append((MovieID, est))
        # Then sort the predictions for each user and retrieve the k highest ones.
    for UserID, ratings in top_n.items():
        ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[UserID] = ratings[:n]

    return top_n
```

```
# First train an SVD algorithm on the movielens dataset.
from surprise import Reader
reader = Reader()
data = Dataset.load_from_df(df[['UserID', 'MovieID', 'Rating']], reader)
trainset = data.build_full_trainset()
algo = SVD()
algo.fit(trainset)
```

```
# Than predict ratings for all pairs (u, i) that are NOT in the training set.
testset = trainset.build_anti_testset()
predictions = algo.test(testset)
```

```
top_n = get_top_n(predictions, n=10)
```

```
# Print the recommended items for each user
for UserID, ratings in top_n.items():
    print(UserID, [MovieID for (MovieID, _) in ratings])
```

***Maths behind the matrix factorisation***

Matrix factorization is a class of collaborative filtering algorithms used in recommender systems. Matrix factorization algorithms work by decomposing the user-item interaction matrix into the product of two lower dimensionality rectangular matrices.

The idea behind matrix factorization is to represent users and items in a lower dimensional latent space.

Matrix factorization can be seen as breaking down a large matrix into a product of smaller ones.


        Rating Matrix[R] = User Matrix[P] * Item Matrix[Q](transpose)

 R(m user * n Items) = P(m user * k features) * Q(k feature * n items)(trans.)

 ***Singular value decomposition ( SVD )***

 It decomposes 
R
 into two unitary matrices and a diagonal matrix:

R
=
U
Σ
V
(Transpose)

where 
R
 is user ratings matrix, 
U
 is the user “features” matrix, 
Σ
 is the diagonal matrix of singular values (essentially weights), and 
V
( Transpose)
 is the movie “features” matrix. 
U
 and 
V
( Transpose )
 are orthogonal, and represent different things. 
U
 represents how much users “like” each feature and 
V
( Transpose )
 represents how relevant each feature is to each movie.

  ***Non-Negative Matrix Factorization ( NMF )***

  NMF decomposes the non-negative utility matrix R into the product of matrices
   W and H: 
R (n ∗ d) = W (n ∗ r) . H (r ∗ d)

Where columns in matrix 
W (n ∗ r)
 represent components, while matrix 
H (r ∗ d)
 stores the corresponding weights. More importantly, NMF introduces constraints under which: 
W
≥
0
 and 
H
≥
0
.The component-wise nonnegativity is a substantial difference from SVD.
